If you like optimization, chances are you've heard of <b style="color:#537FBF">convex optimization</b>, the study of optimization of <b style="color:#A6A440">convex</b> functions.  In this post, we'll look at what it means to be <b style="color:#A6A440">convex</b>, and why we should care.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100)**2)
ax.set_title("$x^2$ - a convex function")
pass
```

A <b style="color:#A6A440">convex function</b> is a function (of potentially multiple variables) with the following property: any line drawn between two points on the function will never pass below the function.  Note that this forces the function to have a single <b style="color:#A6A440">local mimimum</b>^[Well, a flat line is <b style="color:#A6A440">convex</b> but it has infinitely many <b style="color:#A6A440">minima</b> - so we should in theory be more precise about our wording.  Note, however that all <b style="color:#A6A440">minima</b> are just as good as any other.] - think about it!^[Another aside; the <b style="color:#A6A440">convex</b> function $e^x$ has no <b style="color:#A6A440">minima</b>, unless we count the <b style="color:#A6A440">minimum</b> at $-\infty$.  The <strong style="color:#C0CF96">real benefit is that we no longer have to worry about getting stuck in suboptimal <b style="color:#A6A440">local mimima</b></strong>.].  If you consider the shape formed by all the points above the function, you will have a <b style="color:#A6A440">convex</b> shape - hence the name.
